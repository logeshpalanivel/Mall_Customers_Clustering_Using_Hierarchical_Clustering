```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Genre']=le.fit_transform(df['Genre'])
```

```python
import scipy.cluster.hierarchy as sch
```

```python
plt.figure(figsize=(16,8))
dendrogram = sch.dendrogram(sch.linkage(df,method = 'ward'))
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean distance')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
hc = AgglomerativeClustering(n_clusters = 5,affinity = 'euclidean',linkage='average')
y_hc = hc.fit_predict(df)
y_hc
```

```python
X = df.iloc[ : ,[3,4]].values
X
```

```python
X = df[['Annual_Income_(k$)','Spending_Score']].values
```

```python
X
```

```python
plt.figure(figsize=(15, 7))
plt.scatter(X[y_hc == 0, 0], X[y_hc == 0, 1], color='yellow',label='cluster 0', s=50)
plt.scatter(X[y_hc == 1, 0], X[y_hc == 1, 1], color='grey', label='cluster 1', s=50)
plt.scatter(X[y_hc == 2, 0], X[y_hc == 2, 1], color='blue', label='cluster 2', s=50)
plt.scatter(X[y_hc == 3, 0], X[y_hc == 3, 1], color='orange', label='cluster 3', s=50)
plt.scatter(X[y_hc == 4, 0], X[y_hc == 4, 1], color='green', label='cluster 4', s=50)
plt.grid(False)
plt.title('Hierarchy clustering')
plt.xlabel('Annual_Income_(k$)')
plt.ylabel('Spending_Score')
plt.legend()
plt.show()
```
